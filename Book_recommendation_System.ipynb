## 1. Popularity base :

```python
# finding number of ratings on books 

rating_count = dataset.groupby('Book-Title').count()['Book-Rating'].reset_index()

#rename column
rating_count.rename(columns={'Book-Rating':'Num_ratings'},inplace = True)
rating_count
```

## 2. Similarity base(collabrative filtering)

```python
# finding the number of ratings done by each user

user_ratings = dataset.groupby('User-ID').count()['Book-Rating'].reset_index()
user_ratings.rename(columns={'Book-Rating':'no_of_ratings'},inplace=True)

# sorting the user on the number of ratings he is done
user_ratings.sort_values(by='no_of_ratings',ascending=False,inplace = True)
user_ratings
```

### here i consider the users which are ratings more 100 books they are reading lots of books and can be cosider in the recommendation 

```python
# find good reads who read more books

good_reads = user_ratings[user_ratings['no_of_ratings']>100]
good_reads
```

```python
# find the list USER_ID of all good reads
x = good_reads['User-ID']
x = list(x)

filter_data = dataset[dataset['User-ID'].isin(x)]
filter_data
```

```python
# find the number of users rate the book

book_ratings = filter_data.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_ratings.rename(columns={'Book-Rating':'Rating_user_count'},inplace=True)

# sort the books by number of rating users
book_ratings.sort_values(by='Rating_user_count',ascending=False,inplace= True)
book_ratings
```

### here i consider the books which are rated by  more 30 users they are good books  and can be cosider in the recommendation 

```python
# find good books 
good_books=book_ratings [book_ratings['Rating_user_count']>=30]
good_books
```

```python
# find names of all good books
y = good_books['Book-Title']
y = list(y)

final_df = filter_data[filter_data['Book-Title'].isin(y)]
final_df
```

```python
pt = final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
pt[pt.index=='1984']
```

```python
def recommend(book_name):
    
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = final_df[final_df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("1984")
```
